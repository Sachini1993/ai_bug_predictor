```python
import git
import radon
import pandas as pd
 # Test gitpython on one of the cloned repos
repo = git.Repo("repos/flask")
commits = list(repo.iter_commits("main",max_count=10))

for c in commits:
    print(c.hexsha[:7], c.message.strip() [:60])
    
```

```python
all_commits=list(repo.iter_commits("main"))
bug_keywords=["fix","bug","patch", "error","issue"]
bug_commits = [c for c in all_commits if any (k in c.message.lower() for k in bug_keywords)]
print(f"Total commits: {len (all_commits)}")
print(f"Bug-related commits: {len (bug_commits)}")
print(f"Bug ratio: {len(bug_commits)/len (all_commits):.1%}")
```
